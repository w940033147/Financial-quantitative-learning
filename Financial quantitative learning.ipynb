```python
import pandas as pd
import numpy as np
from jqfactor import winsorize
new_df = winsorize(new_df, qrange=[0.05,0.93], inclusive=True, inf2nan=True, axis=0)
```

```python
new_df.isnull().sum()
```

```python
factors = pd.DataFrame(new_df)
factors['date']=factors.index
factors['date']=pd.to_datetime(factors['date'])
factors
```

```python
new_df['WorldQuant_alpha026'] = new_df['WorldQuant_alpha026'].fillna(new_df['WorldQuant_alpha026'].median())
new_df['WorldQuant_alpha084'] = new_df['WorldQuant_alpha084'].fillna(new_df['WorldQuant_alpha084'].median())
new_df['WorldQuant_alpha045'] = new_df['WorldQuant_alpha045'].fillna(new_df['WorldQuant_alpha045'].median())
```

```python
fall = pd.merge(factors,rets,left_on = ['date'],right_on = ['date'],how = 'outer')
fall = fall.set_index('date')
fall
```

```python
icall = fall.apply(lambda x:x.corr(fall['close']),axis = 0)
```

```python
icall = pd.DataFrame(icall)
abs_call = icall.apply(lambda x: abs(x),axis =1)
rank_factor = abs_call.sort_values(by = [0],ascending = False)
```

```python
ff = rank_factor.iloc[1:16,:]
best_factors = list(ff.index.values)
best_factors
```
